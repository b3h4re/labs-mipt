```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("411.ods", engine="odf")
data
```

```python
data_1_5 = data.iloc[11:16, 3:4]
data_1_5 = data_1_5.reset_index(drop=True)
data_1_5 = data_1_5.rename(columns={"ЧАСТЬ 1": "f"})
data_1_5
```

```python
print(f"f_среднее = {np.mean(data_1_5)}")

data_1_5

sigma_f_3 = np.sqrt(np.sum((data_1_5.to_numpy() - np.mean(data_1_5)) ** 2)  / data_1_5.shape[0] / (data_1_5.shape[0] - 1) )
print(f"sigma_f = {sigma_f_3:.2f}")
```
