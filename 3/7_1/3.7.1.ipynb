```python
def MNK(ax,x,y, c = None, m = None) :
    xy = x*y
    if c == None:
        c = 'orange'
    if m == None:
        m = min(x)
    a = (np.mean(xy) - np.mean(x)*np.mean(y))/(np.mean(x**2) - np.mean(x)**2)
    b = -(np.mean(xy)*np.mean(x) - np.mean(x)**2*np.mean(y))/(np.mean(x**2) - np.mean(x)**2) + np.mean(y)
    print(a)
    print(b)
#     print(b)
#     print(xy)
    sigma_a = (1/(len(x)-2)*((np.mean((y - np.mean(y))**2))/ (np.mean( (x - np.mean(x))**2) ) - a**2) )**0.5
    sigma_b = sigma_a * (np.mean(x**2))**0.5
    print(r'$\sigma_a$', sigma_a)
    print(r'$\sigma_b$', sigma_b)
    x = np.arange(m, max(x), 1)
    ax.plot(x,x*a+b, c = c, linewidth=2, label = "k ="+("{:.3f}".format(a)) + ("{:.2f}".format(b)))
#     ax.tick_params(axis='x',  labelsize=20)
#     ax.tick_params(axis='y',  labelsize=20)
    ax.legend(loc='upper left', fontsize=20)
    print("epsilon: ", sigma_a/a)
    print("epsilon: ", sigma_b/b)
    return a, b, sigma_a, sigma_b

def MNK_1(ax,x,y, c = None, m = None) :
    xy = x*y
    if c == None:
        c = 'orange'
    if m == None:
        m = min(x)
    a = (np.mean(xy) - np.mean(x)*np.mean(y))/(np.mean(x**2) - np.mean(x)**2)
    b = -(np.mean(xy)*np.mean(x) - np.mean(x)**2*np.mean(y))/(np.mean(x**2) - np.mean(x)**2) + np.mean(y)
    print(a)
    print(b)
#     print(b)
#     print(xy)
    sigma_a = (1/(len(x)-2)*((np.mean((y - np.mean(y))**2))/ (np.mean( (x - np.mean(x))**2) ) - a**2) )**0.5
    sigma_b = sigma_a * (np.mean(x**2))**0.5
    print(r'$\sigma_a$', sigma_a)
    print(r'$\sigma_b$', sigma_b)
    x = np.arange(m, max(x), 0.05)
    ax.plot(x,x*a+b, c = c, linewidth=2, label = "k ="+("{:.2f}".format(a))+ "+" + ("{:.2f}".format(b)) )
#     ax.tick_params(axis='x',  labelsize=20)
#     ax.tick_params(axis='y',  labelsize=20)
    ax.legend( fontsize=20)
    print("epsilon: ", sigma_a/a)
    print("epsilon: ", sigma_b/b)
    return [a, b]


def MNK_0(ax,x,y, c = None, minim = None) :
    xy = x*y
    if c == None:
        c = 'orange'
    if minim == None:
        minim = min(x)
    a = (np.mean(xy))/(np.mean(x**2) )
    print(a)
#     print(b)
#     print(xy)
    sigma_a = (1/(len(x)-1) * (np.mean(y**2)/np.mean(x**2) - a**2))**0.5
    print(r'$\sigma_a$', sigma_a)
#   print(r'$\sigma_b$', sigma_b)
    x = np.arange(minim, max(x), 1)
    ax.plot(x,x*a, c = c, linewidth=2, label = "k ="+("{:.2f}".format(a*10**5)) + r"$\cdot 10^5$")
#     ax.tick_params(axis='x',  labelsize=20)
#     ax.tick_params(axis='y',  labelsize=20)
    ax.legend(fontsize=15)
    print("epsilon: ", sigma_a/a)
    return a
    
def make_array(s):
    s = s.replace(',','.')
    s = [float(x) for x in s.split('	')]
    s = np.array(s)
    return s
```
